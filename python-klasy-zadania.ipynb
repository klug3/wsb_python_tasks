```python
class Prostokat:
    def __init__(self, a, b):
        self.a = a
        self.b = b
    
    def __str__(self):
        return f"{__class__.__name__}(a={self.a}, b={self.b})"

    def __eq__(self, _obj: object):
        if type(_obj) != type(self):
            return False

        if self.a == _obj.a and self.b == _obj.b:
            return True
        elif self.a == _obj.b and self.b == _obj.a:
            return True
        else:
            return False
            
    def __lt__(self, _obj):
        if type(_obj) != type(self):
            return False
        
        if self.pole() < _obj.pole():
            return True
        else:
            return False
        
    def __gt__(self, _obj):
        if type(_obj) != type(self):
            return False
        else:
            return not (self.__lt__(_obj))
    
    def pole(self):
        return self.a * self.b

    def obwod(self):
        return 2 * (self.a + self.b)
    
    

figura = Prostokat(10, 50)
figura2 = Prostokat(11, 50)

print(f"{figura}: pole={figura.pole()}")

print(figura == figura2)

print(figura < figura2)

print(figura > figura2)
```

```python
class Ulamek:
    '''
    To jest klasa Ulamek obsługująca podstawowe
    działania na ułamkach zwykłych.
    '''
    def __init__(self, num, den) -> None:
        '''
        Konstruktor

        Parameters
        ----------
        num : int
            licznik ułamka
        den : int
            mianownik ułamka

        Returns
        ----------
        None
        '''
        if den == 0:
            raise Exception("ERROR! Mianownik nie moze być równy 0!")
        elif type(num) != int:
            raise Exception(f"ERROR! Licznik musi być liczbą całkowitą (int).")
        elif type(den) != int:
            raise Exception("ERROR! Mianownik musi być liczbą całkowitą (int).")
        
        self.num = num
        self.den = den 

    
    def __str__(self) -> str:
        '''
        Metoda do wyświetlania ułamków w postaci ułamków właściwych,
        albo liczb mieszanych (w przypadku ułamków niewłaściwich).

        Returns
        ----------
        string
            zawierający odpowiednią reprezentację ułamka
        '''
        if self.num / self.den > 1:
            self.floor = self.num // self.den
            self.rest = self.num % self.den
            return f"{self.floor} {self.rest}/{self.den}"
        else:
            return f"{self.num}/{self.den}"
    
    def __eq__(self, _obj: object) -> bool:
        ''''
        Metoda do porównywania ułamków.
        Jezeli porównywane ułamki są równe zwrócona zostaje wartość True.
        
        Parameters
        ----------
        _obj : object
            inny objekt klasy Ulamek do porównania

        Returns
        ---------
        bool
            wynik porównania - True / False
        '''
        if type(_obj) != type(self):
            return False
        
        if (self.num / self.den) == (_obj.num / _obj.den):
            return True
        else:
            return False
        
    def __lt__(self, _obj: object) -> bool:
        '''
        Metoda do porównywania ułamków.
        Jezeli ułamek jest mniejszy od ułamka, z którym jest porównywany zwaracana zostaje wartość True.

        Parameters
        ----------
        _obj : object
            inny obiekt klasy Ulamek do porównania

        Returns
        ----------
        bool
            wynik porównania - True / False
        '''
        if type(_obj) != type(self):
            return False
        
        if (self.num / self.den) < (_obj.num / _obj.den):
            return True
        else:
            return False
        
    def __gt__(self, _obj: object) -> bool:
        '''
        Metoda do porównywannia ułamków.
        Jezeli ułamek jest większy od ułameka, z którym jest porównywany zwaracana zostaje wartość True.
        Wywołuje metodę __lt__ i zwraca wartość przeciwną.

        Parameters
        ----------
        _obj : object
            inny obiekt klasy Ulamek do porównania

        Returns
        ----------
        bool
            wynik porównania - True / False

        '''
        if type(_obj) != type(self):
            return False
        else:
            return not (self.__lt__(_obj))
        
    def __add__(self, _obj: object) -> object:
        '''
        Metoda do dodawania ułamków.
        Korzysta z metod:
        __gcd__ do wyznaczenia najwiekszego wspólnego dzielnika;
        __lcm__ do wyznaczenia najmnijeszego wspólnego mianownika.

        Parameters
        ----------
        _obj : object
            inny obiek klsay Ulamek do porównania
        
        Returns
        ----------
        object
            wynik dodawania ułamków - nowy obiekt klas Ulamek
        '''
        self.lcd = self.__lcm__(self.den, _obj.den)
        self.num1 = self.num * (self.lcd / self.den)
        self.num2 = _obj.num * (self.lcd / _obj.den)
        self.res_num = self.num1 + self.num2
        return Ulamek(int(self.res_num), int(self.lcd))
    
    def __mul__(self, _obj) -> object:
        '''
        Metoda do mnozenia ułamków.

        Parameters
        ----------
        _obj : object
            inny obiekt klasy Ulamek

        Returns
        ----------
        object
            wynik mnozenia ułamków - nowy obiekt klasy Ulamek
        '''
        self.new_num = self.num * _obj.num
        self.new_den = self.den * _obj.den
        return Ulamek(self.new_num, self.new_den)
    
    def __gcd__(self, a, b):
        '''
        Metoda do wyznaczania największego wspólnego dzielnika dwóch liczb.

        Parameters
        ----------
        a : int
            pierwsza liczba
        b : int
            druga liczba

        Returns
        ----------
        int
            największy wspolny dzielnik
        '''
        if (a == 0):
            return b
        else:
            return self.__gcd__(b % a, a)
        
    def __lcm__(self, a, b):
        '''
        Metoda do wyznaczania najmnieszego wspólnego mianownika dwóch ułamków zwykłych.
        Korzysta z metody:
        __gcd__ do wyznaczenia największego wspólnego dzielnika dwóch liczb.

        Parameters
        ----------
        a : int
            mianownik pierwszego ułamka zwykłego
        b : int
            mianownik drugiego ułamka zwykłego

        Returns
        ----------
        int
            największy wspólny mianownik dwóch ułamków zwykłych
        '''
        return (a * b) / self.__gcd__(a, b)
    
    def skracaj(self) -> object:
        '''
        Metoda do skracania ułamków zwykłych.
        Korzysta z metody:
        __gcd__ do wyznaczenia najwiekszego wspólnego dzielnika dla licznika i mianownika.

        Parameters
        ----------
        None
        
        Returns
        ----------
        object
            skrócony ułamek zwykły - nowy obiekt klasy Ulamek
        '''
        self.gc_divisor = self.__gcd__(self.num, self.den)
        self.new_num = int(self.num / self.gc_divisor)
        self.new_den = int(self.den / self.gc_divisor)
        return Ulamek(self.new_num, self.new_den)

    
u1 = Ulamek(1, 6)
u2 = Ulamek(7, 15)

u3 = Ulamek(15, 7)

print(u1 < u2)


print(u1)
print(u2)
print(u3)

u4 = u1 + u2
print(f"{u1} + {u2} = {u3}")

u5 = u1 * u2
print(f"{u1} * {u2} = {u5}")

u6 = Ulamek(5, 10)
print(u6)

u7 = u6.skracaj()
print(u7)

u8 = Ulamek(20, 40)

(print())
print(f"{u8} -> {u8.skracaj()}")

help(Ulamek.skracaj)
```

## Kod źródłowy dla metody dodawania dwóch ułamków zwykłych.

```python
# python Program to determine
# LCD of two fractions and
# Perform addition on fractions

# function to calculate gcd
# or hcf of two numbers.
def gcd(a, b):
	
	if (a == 0):
		return b
	return gcd(b % a, a)


# function to calculate
# lcm of two numbers.
def lcm(a, b):
	
	return (a * b) / gcd(a, b)


def printSum(num1, den1,
				num2, den2):
				
	# least common multiple
	# of denominators LCD
	# of 6 and 15 is 30.
	lcd = lcm(den1, den2);
	
	# Computing the numerators
	# for LCD: Writing 1/6 as
	# 5/30 and 7/15 as 14/30
	num1 *= (lcd / den1)
	num2 *= (lcd / den2)

	# Our sum is going to be
	# res_num/lcd
	res_num = num1 + num2;
	print( int(res_num) , "/" ,
					int(lcd))

# Driver Code
# First fraction is 1/6
num1 = 7
den1 = 15

# Second fraction is 7/15
num2 = 1
den2 = 6
printSum(num1, den1, num2, den2);

# This code is contributed
# by Sam007
```
