```python
class Ulamek:
    def __init__(self, num, den) -> None:
        self.num = num
        self.den = den
    
    def __str__(self) -> str:
        if self.num / self.den > 1:
            self.floor = self.num // self.den
            self.rest = self.num % self.den
            return f"{self.floor} {self.rest}/{self.den}"
        else:
            return f"{self.num}/{self.den}"
    
    def __eq__(self, _obj: object) -> bool:
        if type(_obj) != type(self):
            return False
        
        if self.num == _obj.num and self.den == _obj.den:
            return True
        else:
            return False
        
    def __lt__(self, _obj: object) -> bool:
        if type(_obj) != type(self):
            return False
        
        if (self.num / self.den) < (_obj.num / _obj.den):
            return True
        else:
            return False
        
    def __gt__(self, _obj: object) -> bool:
        if type(_obj) != type(self):
            return False
        else:
            return not (self.__lt__(_obj))
        
    def __add__(self, _obj: object) -> object:
        self.lcd = self.lcm(self.den, _obj.den)
        self.num1 = self.num * (self.lcd / self.den)
        self.num2 = _obj.num * (self.lcd / _obj.den)
        self.res_num = self.num1 + self.num2
        return Ulamek(int(self.res_num), int(self.lcd))
    
    def __mul__(self, _obj) -> object:
        self.new_num = self.num * _obj.num
        self.new_den = self.den * _obj.den
        return Ulamek(self.new_num, self.new_den)
    
    def gcd(self, a, b):
        if (a == 0):
            return b
        else:
            return self.gcd(b % a, a)
        
    def lcm(self, a, b):
        return (a * b) / self.gcd(a, b)
    
    def skracaj(self) -> object:
        self.gc_divisor = self.gcd(self.num, self.den)
        self.new_num = int(self.num / self.gc_divisor)
        self.new_den = int(self.den / self.gc_divisor)
        return Ulamek(self.new_num, self.new_den)

    
u1 = Ulamek(1, 6)
u2 = Ulamek(7, 15)   # 19/30

u3 = Ulamek(15, 7)

print(u1)
print(u2)
print(u3)

u4 = u1 + u2
print(f"{u1} + {u2} = {u3}")

u5 = u1 * u2
print(f"{u1} * {u2} = {u5}")

u6 = Ulamek(5, 10)
print(u6)

u7 = u6.skracaj()
print(u7)
```

```python
# python Program to determine
# LCD of two fractions and
# Perform addition on fractions

# function to calculate gcd
# or hcf of two numbers.
def gcd(a, b):
	
	if (a == 0):
		return b
	return gcd(b % a, a)


# function to calculate
# lcm of two numbers.
def lcm(a, b):
	
	return (a * b) / gcd(a, b)


def printSum(num1, den1,
				num2, den2):
				
	# least common multiple
	# of denominators LCD
	# of 6 and 15 is 30.
	lcd = lcm(den1, den2);
	
	# Computing the numerators
	# for LCD: Writing 1/6 as
	# 5/30 and 7/15 as 14/30
	num1 *= (lcd / den1)
	num2 *= (lcd / den2)

	# Our sum is going to be
	# res_num/lcd
	res_num = num1 + num2;
	print( int(res_num) , "/" ,
					int(lcd))

# Driver Code
# First fraction is 1/6
num1 = 7
den1 = 15

# Second fraction is 7/15
num2 = 1
den2 = 6
printSum(num1, den1, num2, den2);

# This code is contributed
# by Sam007
```

Funkcja make_noise() moze wyrzucic kilka typow wyjatkow (w sposob losowy). Napisz kod, w zaznaczonym miejscu, ktory wywola funkcje make_noise() i przechwyci  wyjatki: ArithmeticError, OverflowError, ZeroDivisionError, TypeError i odpowiednio je obsluzy wypisujac na ekran odpowiedni komunikat:
* ArithmeticError: wypisze na ekranie: '1:' i tresc bledu przekazanego przez wyjatek ( "1: Error: Blad operacji arytmetycznej")
* OverflowError:  wypisz na ekranie:  '2:' i tresc bledu  
* ZeroDivisionError:  wypisze na ekranie: '3:' i tresc bledu
* TypeError: wypisze na ekranie: '4:' i tresc bledu

W przypadku zakonczenia wykonania z sukcesem (bez wyjatku) wypisze na ekranie komunikat: "Sukces! Brak bledow"

```python
import random

wyjatki = [
    ArithmeticError("Error: Blad operacji arytmetycznej"),
    OverflowError("Error: Przepelnienie bufora!"),
    ZeroDivisionError("Error: dzielenie przez zero"),
    TypeError("Error: Niezgodny typ"),
    None
]

def make_noise():
    w = random.choice(wyjatki)
    if w:
        raise w 


# ------------- dodaj swoj kod tutaj --------------
try:
    make_noise()
    print("Sukces! Brak błędów")
except Exception as e:
    print(e)
    # lub z ładniejszym formatowaniem
    # import traceback
    # traceback.print_exc()
    

# ^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^    
    
    
    
```
