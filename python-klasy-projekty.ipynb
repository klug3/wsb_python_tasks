## Projekt 1: gra wisielec
Napisac wykorzystujac programowanie obiektowe prosta gre "Wisielec".
- Jako argumenty startowe gracz podaje: slowo do odgadniecia oraz ilosc zyc (blednych odpowiedzi)
- gra toczy sie do momentu odgadniecia calego slowa lub utraty wszystkich zyc
- aktualny stan gry wyswitlamy w maksymalnie uproszczony sposob: 
  - w odgadywanym slowie pokazujemy odkryte juz literki. Pozostale prezentujemy jako - , dzieki czemu gracz wie ile jeszcze literek pozostalo do odkrycia
  - ilosc popelnionych bledow oraz pozostalych ruchow rowniez powinna byc wyswietlona na ekranie
  - przykladowy wyglad stanu gry:  `slowo=la--a  zycia=[xx...]`
- Jesli podajemy literke i wystepuje ona wielokrotnie - odslaniamy wszystkie wystapienia
- Jesli podamy ta sama literke drugi raz - tracimy jedno zycie
- wielkosc podawanych liter nie ma znaczenia: l i L liczy sie jako taka sama literka
- Jesli podamy znak inny niz literka - tracimy jedno zycie

Klasa Wisielec powinna miec metode `koniec()` sprawdzajaca czy gra dobiegla konca. Jako rezultat metoda ta zwraca True (gra sie skonczyla) albo False (gra sie ciagle toczy).
Klasa Wisielec powinna miec metode `runda()` ktora pozwoli na rozegranie jednej rundy gry (zapyta o literka, sprawdzi ja w slowie, wysietli odpowiedni stan gry). 
Pozostale metody w klasie - wg. uznania autora.

Gre chcemy uruchamiac w nastepujacy sposob:

    gra = Wisielec( "lampa", 10 )
    while not gra.koniec():
        gra.runda()

#### Podpowiedz:
Zeby zapytac uzytkownika o podanie literki mozesz uzyc:
c = input("Podaj literke:")
wprowadzona przez uzytkownika literka zostanie zachowana w zmiennej c.


Wykorzystaj ponizszy kod do testowania swojej gry.

Przykladowy przebieg gry:
```
    slowo=-----  zycie=[...]
    Podaj literke:l

    slowo=l----  zycie=[...]
    Podaj literke:a

    slowo=la--a  zycie=[...]
    Podaj literke:x

    slowo=la--a  zycie=[x..]
    Podaj literke:3
    To nie jest literka! Tracisz zycie.

    slowo=la--a  zycie=[xx.]
    Podaj literke:k
    Przykro mi, przegralas...sprobuj ponownie


````python
# tutaj zakoduj swoja klase
class Wisielec:
    def __init__(self, word, lives) -> None:
        self.word = word.lower()
        self.lives = lives
        self.known_letters = []
        self.lost_lives = 0
        self.displayed_word = ""

    def runda(self):
        self.dislayed_lives = self.create_displayed_lives()
        self.displayed_word = self.create_displayed_word()
        print(f"słowo={self.displayed_word} zycie=[{self.displayed_lives}]")
        self.c = input("Podaj literkę: ").lower()
        self.zweryfikuj_literke(self.c)

    def koniec(self):
        if self.word == self.displayed_word:
            print(f"Gratulacje! Odgadłaś całe słowo: {self.word}!")
            return True
        elif self.lost_lives > self.lives - 1:
            print("Przykro mi, przegrałaś... spróbuj ponownie.")
            return True

    def zweryfikuj_literke(self, litera):
        try:
            litera = int(litera)
            print("To nie jest literka. Tracisz  życie.")
            self.lost_lives += 1
            return
        except:
            pass

        if litera == "":
            print("To nie jest literka. Tracisz życie.")
            self.lost_lives += 1
        elif litera in self.word and litera not in self.known_letters:
            self.known_letters.append(litera)
            self.create_displayed_word()
        elif litera in self.known_letters:
            print("Już podałaś tę literkę!")
        else:
            print("Podałaś niewłaściwą literę!")
            self.lost_lives += 1

    def create_displayed_word(self):
        self.displayed_word = ""
        for character in self.word:
            if character in self.known_letters:
                self.displayed_word += character
            else:
                self.displayed_word += "-"
        return self.displayed_word

    def create_displayed_lives(self):
        self.displayed_lives = self.lost_lives * "x"
        self.displayed_lives += (self.lives - self.lost_lives) * "."
        return self.displayed_lives


# test
gra = Wisielec("lampa", 3)
while not gra.koniec():
    gra.runda()
````
